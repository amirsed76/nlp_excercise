# 8- Stemming

```python
from nltk.stem import PorterStemmer,LancasterStemmer
```

```python
tokens = TreebankWordTokenizer().tokenize(english_text)
```

```python
print("token        PorterStemmer     LancasterStemmer \n_________________________________________")

for index in [2,10,18,19,21,22,42]:
    print(f"{tokens[index].ljust(10)}   {PorterStemmer().stem(tokens[index]).ljust(10)}        {LancasterStemmer().stem(tokens[index])}" )
```

# 9-‫‪Lemmatization‬‬

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer 
```

## 9-1:‫‪WordNetLemmatizer‬‬

```python
print("word        lemmatezed\n___________________________")
for word in ["ate","better","corpora","was","rocks","fishing","waves"]:
    print(f"{word.ljust(10)}  {WordNetLemmatizer().lemmatize(word)}")
```

## 9-2:improve ‫‪Lemmatization‬‬

```python
from nltk.corpus import wordnet
```

```python
print("word        lemmatezed\n___________________________")
for word,pos in [("ate",wordnet.VERB),("better",wordnet.ADJ),("corpora",wordnet.NOUN),("was",wordnet.VERB),("rocks",wordnet.NOUN),("fishing",wordnet.VERB),("waves",wordnet.NOUN)]:
    print(f"{word.ljust(10)}  {WordNetLemmatizer().lemmatize(word , pos=pos)}")
```
