## part c: word decomposition 

```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(pattern="\w+")
```

```python
new_text=""
for sentence in sentences:
    new_text += "<s> "
    new_text += " ".join(word for word in tokenizer.tokenize(sentence))
    new_text += " </s> "
    
```

```python
import pandas as pd
```

```python
def n_gram_probability(raw , unigrams_df , bigrams_df , trigrams_df,quadrigrams_df):
    n_gram = raw["n_gram"]
    V = len(unigrams1_df) #tokens
    if len(n_gram) == 1:
        N = unigrams1_df["count"].sum()
        
    if len(n_gram) == 2:
        # amount of repite n_gram in unigrams
        N = unigrams_df.loc[unigrams_df['n_gram'] == (n_gram[0],)]["count"].values[0]
        
    if len(n_gram) == 3:
        # amount of repite n_gram in unigrams
        N = bigrams_df.loc[bigrams_df['n_gram'] == (n_gram[0],n_gram[1])]["count"].values[0]
        
    if len(n_gram) == 4:
        # amount of repite n_gram in unigrams
        N = trigrams_df.loc[trigrams_df['n_gram'] == (n_gram[0],n_gram[1],n_gram[2])]["count"].values[0]
        
    
    return (raw["count"]+1)/(V+N)
```

## part a : predict word

```python
def predict_word(sequence , n_grams_df):
    df=n_grams_df.copy()
    df["is_contain"]=df.apply(lambda item : item['n_gram'][0:-1] == sequence ,axis=1)
    resulte= df.loc[df["is_contain"] == True]
    resulte.sort_values(by=["probability"],ascending=False)
    try:
        return resulte["n_gram"].values[0][-1]
    except:
        return "__"

    
```

## for with stop_words structure

```python
for n_gram,model in [(("william",),bigrams1_df) ,(("tattered",),bigrams1_df) , 
                    (("ragged", "hand"),trigrams1_df) ,(("sweetly", "chide"),trigrams1_df) ,
                    (("beguile", "the", "world"),quadrigrams1_df) ,(("in", "thy", "noon"),quadrigrams1_df) ,
                    ]:
    print(str(n_gram).ljust(35) , predict_word(n_gram,model))
```

## for without stop_words structure

```python
for n_gram,model in [(("william",),bigrams2_df) ,(("tattered",),bigrams2_df) , 
                    (("ragged", "hand"),trigrams2_df) ,(("sweetly", "chide"),trigrams2_df) ,
                    (("beguile", "the", "world"),quadrigrams2_df) ,(("in", "thy", "noon"),quadrigrams2_df) ,
                    ]:
    print(str(n_gram).ljust(35) , predict_word(n_gram,model))
```

```python
def make_sentence(n_grams,size):
    n_gram_size = len(n_grams[0])
    if len(n_grams) == 0:
        return None
    first_n_gram = [item for item in n_grams if item[0]=="<s>"][0]
    if size < len(first_n_gram):
        return " ".join(word for word in first_n_gram[0:size])
    tokens=list(first_n_gram)
    for iteration in range(len(n_grams)):
        tokens_len = len(tokens)
        for n_gram in n_grams[1:]:
            if tuple(tokens[-(n_gram_size-1):]) == n_gram[0:n_gram_size-1]:
                tokens.append(n_gram[-1])
                break
        if len(tokens) == tokens_len:
            break


            
    text = " ".join(item for item in tokens)
           
    for sentence in RegexpTokenizer("<s>.[\w ]+.</s>").tokenize(new_text):
        if len(RegexpTokenizer("\w+").tokenize(sentence))==size+2:
               return sentence
    return None            
            
        
        
    
```
